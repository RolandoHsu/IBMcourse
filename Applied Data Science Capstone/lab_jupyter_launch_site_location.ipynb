```python
# 創建一個地圖，並設置初始視圖和縮放級別
nasa_coordinate = [29.559684888503615, -95.0830971930759]
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# 創建一個 MarkerCluster 對象
marker_cluster = MarkerCluster().add_to(site_map)

# 遍歷 spacex_df 資料框中的每一行，並根據座標和顏色添加標記到 marker_cluster
for index, record in spacex_df.iterrows():
    # 創建一個 Marker 對象，並根據 class 狀態自定義圖標的顏色
    marker = folium.Marker(
        location=[record['Lat'], record['Long']],  # 發射場的經緯度
        icon=folium.Icon(color='white', icon_color=record['marker_color']),  # 根據 marker_color 設置顏色
        popup=folium.Popup(f"Launch Site: {record['Launch Site']}<br>Class: {record['class']}", max_width=200)  # 顯示發射場和 class
    )
    # 將 Marker 添加到 marker_cluster
    marker_cluster.add_child(marker)

# 將 marker_cluster 添加到地圖
site_map.add_child(marker_cluster)

# 顯示地圖
site_map
```

Next, we need to explore and analyze the proximities of launch sites.


Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

Now zoom in to a launch site and explore its proximity to see if you can easily find any railway, highway, coastline, etc. Move your mouse to these points and mark down their coordinates (shown on the top-left) in order to the distance to the launch site.

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
from math import sin, cos, sqrt, atan2, radians
import folium
from folium import DivIcon

# 定義發射場的經緯度
launch_site_lat = 28.563197
launch_site_lon = -80.576820

# 假設最近城市的經緯度（可以使用 MousePosition 來查找）
city_lat = 28.401
city_lon = -80.605

# 計算距離
distance_to_city = calculate_distance(launch_site_lat, launch_site_lon, city_lat, city_lon)

# 在地圖上標記最近的城市，並顯示距離
city_marker = folium.Marker(
    location=[city_lat, city_lon],
    icon=DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_to_city),  # 顯示距離
    )
)

# 創建一條連接發射場和城市的線
coordinates = [(launch_site_lat, launch_site_lon), (city_lat, city_lon)]
city_line = folium.PolyLine(locations=coordinates, weight=1)

# 將標記和線條添加到地圖上
site_map.add_child(city_marker)
site_map.add_child(city_line)

# 顯示地圖
site_map
```
